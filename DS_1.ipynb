```python
# 1.Import Required libraries

import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  
import seaborn as sns  
import warnings
warnings.filterwarnings("ignore") 
```

```python
# 2.Import the dataset

link = "I:\Datasets\laptopData.csv"
```

```python
# 3.Load the Dataset into pandas dataframe.  

df = pd.read_csv(link)
df
```

```python
# 4.Data Preprocessing

df.isnull().sum()
```

```python
# Replace missing values in numerical columns with their mean
df.fillna(df.select_dtypes(include=['float64']).mean(), inplace=True)

# Replace missing values in categorical columns with their mode (most frequent value)
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Verify if all missing values are handled
df.isnull().sum()
```

```python
#Statistical Summary

df.describe()
```

```python
# Get basic statistics of categorical columns
df.describe(include=['O']) ## 'O' means object type (categorical data)
```

```python
# 5.Data Formatting & Normalization

# Check data types of all columns
print("\nData types of each column:")
print(df.dtypes)

# Convert data types if needed (Example: Convert Price column to float)
df['Price'] = df['Price'].astype(float)  # Uncomment if needed

# Check again after conversion
print("\nUpdated data types:")
df.dtypes
```

```python
# Convert Categorical Variables into Quantitative Variables

from sklearn.preprocessing import LabelEncoder

# Identify categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
print("Categorical Columns:", categorical_columns)

# Apply Label Encoding to each categorical column
encoder = LabelEncoder()
for col in categorical_columns:
    df[col] = encoder.fit_transform(df[col])

# Display the updated DataFrame
df.head()
```
