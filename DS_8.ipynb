```python
import pandas as pd 
import seaborn as sns  # For visualization
import matplotlib.pyplot as plt  # For additional plotting options
import warnings
warnings.filterwarnings("ignore")  # Suppress all warnings
```

```python
# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Display first few rows
titanic.head()
```

```python
# Display dataset information
titanic.info()

# Check for missing values
print("\nMissing Values:\n", titanic.isnull().sum())
```

```python
# (A) Countplot - Number of Passengers by Class

plt.figure(figsize=(6,4))  # Set figure size
sns.countplot(x='class', data=titanic, palette='pastel')  # Create countplot
plt.title('Number of Passengers in Each Class')  # Add title
plt.show()
```

```python
# (B) Gender Distribution of Survivors

plt.figure(figsize=(6,4))
sns.countplot(x='sex', hue='survived', data=titanic, palette='Set2')
plt.title('Survival Count by Gender')
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
# (C) Boxplot – Ticket Fare by Passenger Class

plt.figure(figsize=(8,5))  # Set figure size
sns.boxplot(x='class', y='fare', data=titanic, palette='coolwarm')  
plt.title('Ticket Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare ($)')
plt.show()
```

```python
# (D) Bar Chart – Survival Rate by Passenger Class

plt.figure(figsize=(6,4))
sns.barplot(x='class', y='survived', data=titanic, palette='viridis', estimator=lambda x: sum(x)/len(x))
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.ylim(0, 1)  # Scale from 0 to 1 (percentage)
plt.show()
```

```python
# (E). Survival Count Plot

sns.countplot(x='survived', data=titanic)
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()
```

```python
# (G).  Survival Rate by Age and Sex (Point Plot)

sns.pointplot(x='age', y='survived', hue='sex', data=titanic)
plt.title("Survival Rate by Age and Sex")
plt.show()
```

```python
# (H) Survival Probability vs Fare (Lineplot)

sns.lineplot(x='fare', y='survived', data=titanic)
plt.title("Survival Probability by Fare")
plt.show()
```

```python
# Plot Histogram of Ticket Prices ('fare')

plt.figure(figsize=(8,5))  # Set figure size
sns.histplot(titanic['fare'], bins=30, kde=True, color='skyblue')  # Create histogram
plt.title('Distribution of Ticket Prices (Fare)')  # Add title
plt.xlabel('Ticket Fare ($)')  # Label X-axis
plt.ylabel('Count')  # Label Y-axis
plt.show()
```
