```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore") 
```

```python
df = pd.read_csv(r"C:\Users\Aditya\Downloads\Income_data.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Statistical Summary for Numeric Variables Grouped by a Categorical Variable
# mean, median, min, max, and standard deviation of numerical columns, grouped by the "Age Group" category

summary_stats = df.groupby("Age Group")[["Income (INR)"]].agg(
    ["mean", "median", "min", "max" ,"std"]
)

summary_stats
```

```python
# Convert Categorical Variable into Numeric List
# Since "Age Group" is categorical, we will assign a numeric value to each unique group.


#Creating a dictionary to map age groups to numbers
age_group_mapping = {"18-25": 1, "26-35": 2, "36-45": 3, "46-60": 4}

# Applying the mapping to create a numeric list
df["Age Group Numeric"] = df["Age Group"].map(age_group_mapping)

# Display the updated dataframe
df.head()
```
